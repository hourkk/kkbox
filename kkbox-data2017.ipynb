```python
import numpy as np
import pandas as pd
```

members_v3一樣 性別留下NAN 年齡挑在0~99

```python
members_v3 = pd.read_csv("members_v3.csv",parse_dates=['registration_init_time'],dtype={'city':np.int8,'bd':np.int8,
                                    'registered_via':np.int8}).query("bd>0 or bd<100")
members_v3['city']=members_v3['city'].astype('category')
members_v3['registered_via']=members_v3['registered_via'].astype('category')
members_v3.head()
```

```python
members_v3.info()
```

因為個人電腦記憶體不夠只取到2017年之後
user_logs取2017之後 統計每個使用者聽歌的總天數 還有每個欄位平均數
EX.('num_25', 'mean') 這用戶平均每天只聽不到25%長度的數量 其他以此類推
新增一個play_days播放次數 代表該使用者共撥放幾天(該使用者在原始資料出現筆數)

```python
user_logs2017 = pd.read_csv('user_logs2017.csv')
user_logs2017.head()
```

```python
user_logs2017.info()
```

transactions一樣取每個使用者交易統計路
mode是眾數 代表該使用者使用最多次的選項 
EX payment_method_id_mode = 41 代表該使用者最多使用的交易方法是41 
要與他們方案做用眾數比較好解釋
sum是總和 代表該使用者該項總和(次數)
多新增一個transactions_times交易次數 代表該使用者交易次數(該使用者在原始資料有多少筆)

```python
transactions2017g = pd.read_csv("transactions2017g.csv")
transactions2017g['payment_method_id_mode']=transactions2017g['payment_method_id_mode'].astype('category')
transactions2017g.head()
```

```python
transactions2017g.info()
```
